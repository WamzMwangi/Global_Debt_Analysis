### Preparing Non-Financial Corporate Debt

## Joining the 4 DataFrames

```python
data=pd.merge(gov_debt, house_debt, right_index=True, left_index=True)
```

```python
data.head()
```

```python
data=pd.merge(data,priv_debt, right_index=True, left_index=True)
```

```python
data.head()
```

```python
data=pd.merge(data,corp_debt, right_index=True, left_index=True)
```

```python
data.head()
```

## Data Analysis & Visualization

```python
data.info()
#No missing data
```

### Overall trend in debt from 1950-2022

```python

plt.figure(figsize=(20,8))

sns.lineplot(data=data, y='Central Government Debt',x=data.index,linewidth=4,label='Central Goverment Debt')
sns.lineplot(data=data, y='Household Debt',x=data.index,lw=4,label='Household Debt')
sns.lineplot(data=data, y='Private Debt',x=data.index,lw=4,label='Private Debt')
sns.lineplot(data=data, y='Non- Financial Corporate Debt',x=data.index,lw=4,label='Corporate NF Debt')

plt.title('Trends in Debt in the United States: 1950-2022',weight='bold',size=20)
plt.ylabel('Annual average of debt (% of GDP)',weight= 'bold',size=14)
plt.xlabel('Year',weight='bold',size=14)
#plt.xticks(rotation=90)
#x axis is still too busy
for index, label in enumerate(plt.gca().get_xticklabels()):
   if index % 5 == 0:
      label.set_visible(True)
   else:
      label.set_visible(False)
plt.legend()
plt.show()
```
