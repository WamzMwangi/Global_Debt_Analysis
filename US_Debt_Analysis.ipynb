## Joining the 4 DataFrames

```python
data=pd.merge(gov_debt, house_debt, right_index=True, left_index=True)
```

```python
data.head()
```

```python
data=pd.merge(data,priv_debt, right_index=True, left_index=True)
```

```python
data.head()
```

```python
data=pd.merge(data,corp_debt, right_index=True, left_index=True)
```

```python
data.head()
```

## Data Analysis & Visualization

```python
data.info()
#No missing data
```

### Overall trend in debt from 1950-2022

```python

plt.figure(figsize=(12,5))

sns.lineplot(data=data, y='Central Government Debt',x=data.index,linewidth=4,label='Central Goverment Debt')
sns.lineplot(data=data, y='Household Debt',x=data.index,lw=4,label='Household Debt')
sns.lineplot(data=data, y='Private Debt',x=data.index,lw=4,label='Private Debt')
sns.lineplot(data=data, y='Non- Financial Corporate Debt',x=data.index,lw=4,label='Corporate NF Debt')

plt.title('Trends in Debt in the United States: 1950-2022',weight='bold',size=20)
plt.ylabel('Annual average of debt (% of GDP)',weight= 'bold',size=14)
plt.xlabel('Year',weight='bold',size=14)
#plt.xticks(rotation=90)
#x axis is still too busy
for index, label in enumerate(plt.gca().get_xticklabels()):
   if index % 5 == 0:
      label.set_visible(True)
   else:
      label.set_visible(False)
plt.savefig(r'C:\Users\User\Desktop\Global Debt Database\Trends in Debt in the United States: 1950-2022.png')
plt.legend()
plt.show()
```

#### Comparison between debt types

```python
data.head()
```

```python
average=data.mean(axis=0)
average
```

```python
sorted_average=average.sort_values(ascending=False)
```

```python
plt.figure(figsize=(15,5))
bars=plt.bar(sorted_average.index, sorted_average.values,color='lightsteelblue')
plt.ylabel('Average Debt (% of GDP)',weight='bold')
plt.title('Total Average Debt over the period of 1950-2022',weight='bold',)
plt.xticks(weight='bold')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 2), va='bottom', weight='bold')
#plt.tight_layout()
plt.show()
```

### Trends in each debt type

```python
# Convert index to numeric values
data.index = pd.to_numeric(data.index)

plt.figure(figsize=(10,6))

# Plotting the line plot
sns.lineplot(data=data, y='Central Government Debt', x=data.index, linewidth=4, label='Central Government Debt')
plt.title('Trend in Central Government Debt: 1950-2022',weight='bold',size=15)
plt.ylabel('Debt to GDP ratio (% of GDP)')
plt.xlabel('YEAR')
# Adding a trendline
sns.regplot(data=data, y='Central Government Debt', x=data.index, scatter=False, color='red',order=1)
plt.show()
```

```python
# Convert index to numeric values
data.index = pd.to_numeric(data.index)
data['Household Debt'] = pd.to_numeric(data['Household Debt'], errors='coerce') 

plt.figure(figsize=(10,6))

# Plotting the line plot
sns.lineplot(data=data, y='Household Debt', x=data.index, linewidth=4, label='Household Debt')
plt.title('Trend in Household Debt: 1950-2022',weight='bold',size=15)
plt.ylabel('Debt to GDP ratio (% of GDP)')
plt.xlabel('YEAR')
# Adding a trendline
sns.regplot(data=data, y='Household Debt', x=data.index, scatter=False, color='red',order=1)
plt.show()
```

```python
# Convert index to numeric values
data.index = pd.to_numeric(data.index)
data['Private Debt'] = pd.to_numeric(data['Private Debt'], errors='coerce') 

plt.figure(figsize=(10,6))

# Plotting the line plot
sns.lineplot(data=data, y='Private Debt', x=data.index, linewidth=4, label='Private Debt')
plt.title('Trend in Private Debt: 1950-2022',weight='bold',size=15)
plt.ylabel('Debt to GDP ratio (% of GDP)')
plt.xlabel('YEAR')
# Adding a trendline
sns.regplot(data=data, y='Private Debt', x=data.index, scatter=False, color='red',order=1)
plt.show()
```
